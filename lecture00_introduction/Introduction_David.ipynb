### Jobs on market

#### A typical bank role

**Expected base salary : 100k~120k**

* Coding skills - SQL, SAS, R
* Banking credit risk knowledge - very important
* Banking product knowledge - important
* Experience  past data analytics technical experience 

**Apart from technical skills, you need basic banking risk & product knowledge:**

E.g. LVR, exposure, collateral, asset-backed, default, delinquency, bad debt, regulatory capital, economic capital, provision, risk grade, PD, LGD, EAD, etc. 

#### A typical marketing data scientist role

**Expected salary : 120~140k**

Coding skills - SQL, R, Python
Customer analytics - very important
Marketing analytics - very important
Experience  past working experience, ML projects, Kaggle competition 

**Some basic customer/marketing analytics:**

Digital data, spending, past purchasing, revenue, acquisition, cross-sell, churn, propensity models, customer lifetime value, NBC/NBA/NBO, segmentation, sentiment analysis, A/B testing, causal inference
